# Predicción

- Objetivo: Generar las predicciones finales sobre el conjunto de datos sin etiqueta utilizando el modelo seleccionado durante la etapa de validación.

# Importación de Librerias

```python
import pandas as pd

from pathlib import Path
import joblib
```

# PATHS

```python
PROJECT_ROOT = Path.cwd().parent


PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"
MODELS_DIR = PROJECT_ROOT / "models"
PREDICTIONS_DIR = PROJECT_ROOT / "predictions"
RAW_DIR = PROJECT_ROOT / "data" / "raw"
```

# Verificación PATHS

```python
print(PROCESSED_DIR.exists())
print(MODELS_DIR.exists())
print(RAW_DIR.exists())
```

# Crear carpeta PREDICTIONS

```python
PREDICTIONS_DIR.mkdir(exist_ok=True)

print("Carpeta predictions lista.")
```

# Cargar datos

```python
test_fe = pd.read_csv(
    PROCESSED_DIR / "test_fe.csv"
)

print(test_fe.shape)
```

# Cargar Modelo

```python
rf = joblib.load(
    MODELS_DIR / "random_forest_final.pkl"
)

print("Modelo cargado.")
```

# Generar predicciones

```python
predicciones = rf.predict(test_fe)

predicciones[:10]
```

# Recuperar Identificadores

- Se carga el DataSet original de entrega para recuperar la columna `ID`, la cual será utilizada para generar el archivo final de predicciones solicitado por la consigna.

```python
entrega_original = pd.read_csv(
    RAW_DIR / "smoking_prediction_entrega.xlsx - smoking_prediction.csv"
)

print(entrega_original.shape)
```

# Verificación de Columnas

- Se verifica la existencia de la columna `ID` necesaria para la generación del archivo final de entrega.

```python
print(entrega_original.columns.tolist())
```

```python
entrega_original[["ID"]].head()
```

# Generación del Archivo Final

- Se combinan los identificadores originales con las predicciones generadas por el modelo Random Forest seleccionado durante la etapa de validación.

```python
resultado_final = pd.DataFrame({
    "ID": entrega_original["ID"],
    "smoking_prediction": predicciones
})

resultado_final.head()
```

# Validación de dimensiones

- Se verifica que la cantidad de registros coincida entre el dataset original de entrega, el conjunto procesado y las predicciones generadas.

```python
print("Dataset original :", len(entrega_original))
print("Dataset procesado:", len(test_fe))
print("Predicciones     :", len(predicciones))
print("Submission       :", len(resultado_final))
```

# Exportación de Predicciones

- Se exporta el archivo final con la estructura requerida para la entrega del TP2.

```python
resultado_final.to_csv(
    PREDICTIONS_DIR / "smoking_prediction.csv",
    index=False
)

print("Archivo exportado correctamente.")
```
